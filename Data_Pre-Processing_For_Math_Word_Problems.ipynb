# **Data Pre-Processing**

```python
nltk.download('wordnet')
```

```python
def synonym_replacement(sentence):
    words = sentence.split()
    new_words = words.copy()
    for i in range(len(words)):
        syns = wordnet.synsets(words[i])
        if syns:
            new_word = syns[0].lemmas()[0].name()
            new_words[i] = new_word if random.random() > 0.5 else words[i]
    return ' '.join(new_words)
```

```python
# Augment the existing data
augmented_data = []
for index, row in df.iterrows():
    augmented_problem = synonym_replacement(row['Question'])
    original_equation = row['Equation']
    augmented_data.append({'Question': augmented_problem, 'Equation': original_equation})

# Convert the augmented data to a DataFrame
augmented_data_df = pd.DataFrame(augmented_data)

# Add the augmented data to the existing DataFrame
combined_data_df = pd.concat([df, augmented_data_df], ignore_index=True)
```

```python
# Find and display the repeating values in a specific column
repeating_values = combined_data_df['Question'][combined_data_df['Question'].duplicated()]
print("Repeating values in 'Question':")
print(repeating_values)
```

```python
#Remove duplicates from the 'Column_Name' column
df = combined_data_df.drop_duplicates(subset=['Question'])
```

```python
repeating_values = df['Question'][df['Question'].duplicated()]
if repeating_values.empty:
    print("No duplicates")
else:
  print('There are duplicates')
```

```python
# Shuffle the dataset
shuffled_df = df.sample(frac=1, random_state=42)  # frac=1 means the whole dataset will be shuffled
```

```python
# Split the dataset into training and testing sets
train_df, test_df = train_test_split(shuffled_df, test_size=0.2, random_state=42)  # 80% training, 20% testing
```

```python
# Save the training and testing sets to separate Excel files
train_df.to_excel('/content/drive/MyDrive/Research_Project/Codes/Train.xlsx', index=False)
test_df.to_excel('/content/drive/MyDrive/Research_Project/Codes/Validation.xlsx', index=False)
```
