```python
def convert_to_gpt35_format(dataset):
    fine_tuning_data = []
    for _, row in dataset.iterrows():
        json_response = '{"Equation": "' + row['Equation'] + '", "Type": "' + row['Type'] + '"}'
        fine_tuning_data.append({
            "messages": [
                {"role": "user", "content": row['Question']},
                {"role": "system", "content": json_response},
                {"role": "assistant", "content": row['Equation']}
            ]
        })
    return fine_tuning_data
```

```python
dataset = pd.read_excel("/content/drive/MyDrive/Research_Project/Codes/New.xlsx")
converted_data = convert_to_gpt35_format(dataset)
converted_data[0]['messages']
```

### **Train-Test split**

```python
from sklearn.model_selection import train_test_split

# Stratified splitting. Assuming 'Top Category' can be used for stratification
train_data, val_data = train_test_split(
    converted_data,
    test_size=0.2,
    stratify=dataset['Type'],
    random_state=42  # for reproducibility
)
```

```python
def write_to_jsonl(data, file_path):
    with open(file_path, 'w') as file:
        for entry in data:
            json.dump(entry, file)
            file.write('\n')

training_file_name = "/content/drive/MyDrive/Research_Project/Codes/JSONGPT3.5/train.jsonl"
validation_file_name = "/content/drive/MyDrive/Research_Project/Codes/JSONGPT3.5/val.jsonl"

write_to_jsonl(train_data, training_file_name)
write_to_jsonl(val_data, validation_file_name)
```
