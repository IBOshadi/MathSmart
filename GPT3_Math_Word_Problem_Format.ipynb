# **Install Libraries**

```python
import json
import pandas as pd
from openai import OpenAI
from sklearn.model_selection import train_test_split
```

# **Format Data**

```python
df = pd.read_excel('/content/drive/MyDrive/Research_Project/Codes/math_word_problems_dataset.xlsx',nrows=2500)
```

```python
df.head()
```

```python
def create_dataset(df):
  fine_tuning_data = []
  for _, row in df.iterrows():
    fine_tuning_data.append (
       {"prompt": row['question'], "completion": row['answer']}
    )
  return fine_tuning_data
```

```python
converted_data = create_dataset(df)
```

```python
train_data, val_data = train_test_split(
    converted_data,
    test_size=0.2,
    random_state=42
)
```

```python
def write_to_jsonl(data, file_path):
    with open(file_path, 'w') as file:
        for entry in data:
            json.dump(entry, file)
            file.write('\n')
```

```python
training_file_name = "/content/drive/MyDrive/Research_Project/Codes/JSON2GPT3/train.jsonl"
validation_file_name = "/content/drive/MyDrive/Research_Project/Codes/JSON2GPT3/val.jsonl"
```

```python
write_to_jsonl(train_data, training_file_name)
write_to_jsonl(val_data, validation_file_name)
```
