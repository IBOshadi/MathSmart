```python
from datasets import load_dataset

dataset = load_dataset("microsoft/orca-math-word-problems-200k")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.DataFrame(dataset['train'])

# Write the DataFrame to an Excel file
df.to_excel('/content/drive/MyDrive/Research_Project/Codes/math_word_problems_dataset.xlsx', index=False)
```

```python
df = pd.read_excel('/content/drive/MyDrive/Research_Project/Codes/math_word_problems_dataset.xlsx')
```

```python
# Print the keys (header) of the dataset
print(dataset['train'].column_names)
```

```python
# Print the first few data points
print(df.head())
```

```python
# Check for missing values in the DataFrame
missing_values = df.isnull().sum()

# Print the count of missing values for each column
print(missing_values)
```

```python
# Print the number of data rows
num_rows = df.shape[0]

print("Number of data rows:", num_rows)
```

```python
# Find duplicate rows
duplicate_rows_df = df[df.duplicated(keep=False)]

# Check if there are any duplicate rows
if not duplicate_rows_df.empty:
   # Print the duplicate rows
   print(duplicate_rows_df)
else:
   # Print a message indicating that there are no duplicates
   print('No duplicates found.')
```

```python
# Drop the duplicate rows
df = df.drop_duplicates(keep='first')
```

```python
# Find duplicate rows based on the combination of equation 1 and equation 2
duplicate_rows_df = df.duplicated(subset=['question', 'answer'], keep=False)

# Check if there are any duplicate rows
if duplicate_rows_df.any():
   # Print the duplicate rows
   print(df[duplicate_rows_df])
else:
   # Print a message indicating that there are no duplicates
   print('No duplicates found.')
```

```python
num_rows1 = df.shape[0]

print("Number of data rows:", num_rows1)
```

```python
# Shuffle the dataset
shuffled_df = df.sample(frac=1, random_state=42)  # frac=1 means the whole dataset will be shuffled
```

```python
# Split the dataset into training and testing sets
train_df, test_df = train_test_split(shuffled_df, test_size=0.2, random_state=42)  # 80% training, 20% testing
```

```python
# Save the training and testing sets to separate csv files
train_df.to_csv('/content/drive/MyDrive/Research_Project/Codes/train_new_data.csv', index=False)
test_df.to_csv('/content/drive/MyDrive/Research_Project/Codes/validation_new_data.csv', index=False)
```
