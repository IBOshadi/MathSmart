# **Fine-tuning**

```python
training_file_name = "/content/drive/MyDrive/Research_Project/Codes/JSON2GPT3/train.jsonl"
validation_file_name = "/content/drive/MyDrive/Research_Project/Codes/JSON2GPT3/val.jsonl"
```

```python
client = OpenAI(api_key="")
```

```python
training_file = client.files.create(
    file=open(training_file_name, "rb"), purpose="fine-tune"
)
validation_file = client.files.create(
    file=open(validation_file_name, "rb"), purpose="fine-tune"
)

print("Training file id:", training_file.id)
print("Validation file id:", validation_file.id)
```

```python
suffix_name = "GPT3"

response = client.fine_tuning.jobs.create(
    training_file=training_file.id,
    validation_file=validation_file.id,
    model="davinci-002",
    suffix=suffix_name,
    hyperparameters={
        "n_epochs": 5,
        "batch_size":1,
        "learning_rate_multiplier":1,
    }

)
response
```

```python
# Assuming 'response' contains the fine-tuning job details
job_id = response.id
job_details = client.fine_tuning.jobs.retrieve(job_id)
print("Job Status:", job_details.status)
```

```python
job_id=""
response = client.fine_tuning.jobs.retrieve(job_id)#job id
response
```

```python
question = "A farmer has 120 acres of land. He plants corn on 3/4 of the land and soybeans on the rest. How many more acres of corn than soybeans does he plant?"
```

```python
FINE_TUNED_MODEL = response.fine_tuned_model
MODEL = "davinci-002"
temperature=1
max_tokens=500

response1 = client.completions.create(
    model=FINE_TUNED_MODEL,
    prompt=question,
    max_tokens=max_tokens,
    temperature=temperature
    )

response2 = client.completions.create(
    model=MODEL,
    prompt=question,
    max_tokens=max_tokens,
    temperature=temperature
    )

print('Question: ',question)
print('*****************************************************************************************')
print('')
print('Answer before fine-tuning: ')
print('')
print(response2.choices[0].text)
print('*****************************************************************************************')
print('')
print('Answer after fine-tuning: ')
print('')
print(response1.choices[0].text)
```
